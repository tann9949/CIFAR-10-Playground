# CIFAR-10 Playground

## Load and formating data

```python
import numpy as np
import matplotlib.pyplot as plt

SEED = 1234
```

```python
idx2class = {
    0:'airplane',
    1:'automobile',
    2:'bird',
    3:'cat',
    4:'deer',
    5:'dog',
    6:'frog',
    7:'horse',
    8:'ship',
    9:'truck'
}
```

```python
import pickle

def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
def feats2img(data):
    r = data[:1024].reshape(32,32)
    g = data[1024:2048].reshape(32,32)
    b = data[2048:].reshape(32,32)
    d = np.dstack([r,g,b])
    return d

def format_data(data):
    out = list()
    size = int((data.shape[1]/3)**0.5)
    for i in range(data.shape[0]):
        feats2img(data[i])
        assert data[i].shape == (32,32,3)
        out.append(data[i])
    return np.array(out)

def get_image_class(idx, X, y):
    x = X[idx]
    if len(x.shape) < 3:
        x = feats2img(x)
    plt.imshow(x)
    plt.title('Class: {}'.format(idx2class[y[idx]]))
    plt.show()
```

```python
def load_cifar():
    X, y = list(), list()
    for i in range(1,6):
        d = unpickle('cifar-10-batches-py/data_batch_{}'.format(i))
        [X.append(x) for x in d[b'data']]
        [y.append(gt) for gt in d[b'labels']]
    return np.array(X), np.array(y)
```

```python
X, y = load_cifar()
```

```python
get_image_class(7, X, y)
```

## Defining PCA Function

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```python
X_svm, y_svm = X[:5000], y[:5000]
X_train, X_test, y_train, y_test = train_test_split(X_svm, y_svm, test_size=0.2, random_state=SEED, stratify=y_svm)
```

```python
%%time
pca_010 = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=int(0.1*X_train.shape[1])))]).fit(X_train)
pca_030 = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=int(0.3*X_train.shape[1])))]).fit(X_train)
pca_050 = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=int(0.5*X_train.shape[1])))]).fit(X_train)
pca_070 = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=int(0.7*X_train.shape[1])))]).fit(X_train)
```

## Run SVM on different PCA scaling

```python
from sklearn.model_selection import KFold
from sklearn.svm import SVC
from sklearn.metrics import f1_score, accuracy_score
```

```python
def score(y_true, y_pred):
    return f1_score(y_true, y_pred, average=None), accuracy_score(y_true, y_pred)
```

```python
import time

def print_f1(f1):
    for i, f in enumerate(f1):
        print('\t{}: {}'.format(idx2class[i], f))

def kfold_SVM(X, y, n_fold=10, seed=SEED, kernel='linear', pca=None):
    if pca:
        X = pca.transform(X)
    kf = KFold(n_splits=n_fold, shuffle=True, random_state=seed)
    avg_f1, avg_acc = list(), list()
    for i, index in enumerate(kf.split(X)):
        start_time = time.time()
        train_index, val_index = index
        X_train, X_val = X[train_index], X[val_index]
        y_train, y_val = y[train_index], y[val_index]
        print('='*20,'Fold {}'.format(i+1), '='*20)
        
        model = SVC(kernel=kernel)
        
        fitted = model.fit(X_train, y_train)
        
        pred_train = model.predict(X_train)
        pred_val = model.predict(X_val)
        
        train_f1, train_acc = score(y_train, pred_train)
        val_f1, val_acc = score(y_val, pred_val)
        
        avg_f1.append(val_f1)
        avg_acc.append(val_acc*100)
        
        print('Finished in {} seconds'.format(round(time.time()-start_time, 2)))
        print('train_accuracy: {}%\tval_accuracy: {}%'.format(round(train_acc*100, 2), round(val_acc*100, 2)))
        print('train_f1:')
        print_f1(train_f1)
        print('val_f1:')
        print_f1(val_f1)
    print('*'*20)
    print('\nAverage F1-Score:')
    print_f1(np.mean(avg_f1, axis=0))
    print('\nAverage Accuracy: {}'.format(round(np.mean(avg_acc), 2)))
    return avg_f1, avg_acc, model
```

```python
def eval_svm(X_test, y_test, model):
    f1, acc = score(y_test, model.predict(X_test))
    print('Test F1-Score:')
    print_f1(f1)
    print('Test Accuracy:')
    print('\t{}%'.format(round(acc*100,2)))
    return f1, acc
```

```python
def plot_f1(f1, xticks, title):
    plt.figure(figsize=(20,7))
    for i in range(f1.shape[1]):
        plt.plot(f1[:,i],'o--', label=idx2class[i])
    plt.legend()
    plt.xticks(np.arange(f1.shape[0]),xticks)
    plt.grid()
    plt.title(title)
    plt.show()
    
def plot_acc(acc, xticks, title):
    plt.figure(figsize=(20,5))
    plt.plot(acc, 'o--')
    plt.xticks(np.arange(acc.shape[0]), xticks)
    plt.title(title)
    plt.grid()
    plt.show()
```

```python
print('>>>>>>>>Training SVM on 10% dimensions PCA...\n')
f1_010_val, acc_010_val, SVC_010 = kfold_SVM(X_train, y_train, pca=pca_010)
print('\n>>>>>>>>Training SVM on 30% dimensions PCA...\n')
f1_030_val, acc_030_val, SVC_030 = kfold_SVM(X_train, y_train, pca=pca_030)
print('\n>>>>>>>>Training SVM on 50% dimensions PCA...\n')
f1_050_val, acc_050_val, SVC_050 = kfold_SVM(X_train, y_train, pca=pca_050)
print('\n>>>>>>>>Training SVM on 70% dimensions PCA...\n')
f1_070_val, acc_070_val, SVC_070 = kfold_SVM(X_train, y_train, pca=pca_070)
print('\n>>>>>>>>Training SVM on 100% dimensions PCA...\n')
f1_linear_val, acc_linear_val, SVC_linear = kfold_SVM(X_train, y_train)
```

```python
X_test_010 = pca_010.transform(X_test)
X_test_030 = pca_030.transform(X_test)
X_test_050 = pca_050.transform(X_test)
X_test_070 = pca_070.transform(X_test)
```

```python
f1_PCA = np.array([f1_010_test, f1_030_test, f1_050_test, f1_070_test, f1_linear_test])
xticks = ['307 component (10% dimensions)','921 components (30% dimensions)','1536 components (50% dimensions)','2150 components (70% dimensions)','3072 components (100% dimensions)']
title = 'per-class F1 Score of different PCA components on linear SVM'
plot_f1(f1_PCA, xticks, title)

acc_PCA = np.array([acc_010_test, acc_030_test, acc_050_test, acc_070_test, acc_linear_test])
title = 'Accuracy of different PCA components on linear SVM'
plot_acc(acc_PCA, xticks, title)
```

## CIFAR-10 using CNN

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data.sampler import SubsetRandomSampler
```

```python
def get_data(cifar_dir, batch_size, augment, validation_split, shuffle=True, num_workers=0, random_seed=SEED):
    
    transform = transforms.Compose(
        [transforms.ToTensor(),
         transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])
    
    if augment:
        train_transform = transforms.Compose([
            transforms.RandomCrop(32, padding=4),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
        ])
    else:
        train_transform = transforms.Compose(
            [transforms.ToTensor(),
             transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
        ])

    train_set = torchvision.datasets.CIFAR10(root=cifar_dir, train=True, download=False, transform=train_transform)
    val_set = torchvision.datasets.CIFAR10(root=cifar_dir, train=True, download=False, transform=transform)
    test_set = torchvision.datasets.CIFAR10(root=cifar_dir, train=False, download=False, transform=transform)

    num_train = len(train_set)
    indices = list(range(num_train))
    split = int(np.floor(validation_split * num_train))
    
    if shuffle:
        np.random.seed(random_seed)
        np.random.shuffle(indices)
        
    train_idx, val_idx = indices[split:], indices[:split]
    train_sampler = SubsetRandomSampler(train_idx)
    val_sampler = SubsetRandomSampler(val_idx)
    
    train_loader = torch.utils.data.DataLoader(
        train_set, batch_size=batch_size, sampler=train_sampler,
        num_workers=num_workers)
    val_loader = torch.utils.data.DataLoader(
        val_set, batch_size=len(val_idx), sampler=val_sampler,
        num_workers=num_workers)
    test_loader = torch.utils.data.DataLoader(
        test_set, batch_size=len(test_set), num_workers=num_workers)
    
    return train_loader, val_loader, test_loader
```

```python
def get_onehot(idx, num_classes=10):
    onehot = torch.zeros(1,10)
    onehot[0][idx] = 1.
    return onehot

def label2onehot(y):
    onehot = None
    for pred in y:
        if onehot is None:
            onehot = get_onehot(pred)
        else:
            onehot = torch.cat([onehot, get_onehot(pred)], dim=0)
    return onehot
```

```python
from sklearn.metrics import f1_score

def get_val_stat(model, val_loader, loss_fn):
    X_val, y_val = iter(val_loader).next()
        
    val_pred = model(X_val)
    val_loss = round(loss_fn(val_pred, y_val).item(), 4)
    val_correct = (y_val == torch.argmax(val_pred, dim=1)).sum().item()
    val_acc = round(val_correct*100/X_val.shape[0], 4)
    val_f1 = f1_score(y_val, torch.argmax(val_pred, axis=1), average='macro')
    
    return val_loss, val_acc, val_f1
```

```python
from sklearn.metrics import f1_score
import time

def train(model, train_loader, val_loader, epochs, loss_fn, batch_size, validation_split=None, lr=1e-3, optimizer='adam', weight_decay=0, momentum=None, verbose=None, seed=SEED):
    model.train()
    step_loss_train, step_acc_train, step_f1_train  = list(), list(), list()
    step_plot = list()
    epoch_loss_val, epoch_acc_val, epoch_f1_val = list(), list(), list()
    epoch_plot = list()
    
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    
    N = np.sum([x[0].shape[0] for x in train_loader])
    num_step = len(train_loader)
    total_correct = 0
    print('Start Training...')
    print('='*10,'Hyperparameters','='*10)
    print('Batch size: {}\nEpochs: {}\nOptimizer: {}\nLearning Rate: {}'.format(batch_size, epochs, optimizer, lr))
    print('Number of Parameters: {}'.format(total_params))
    print('Number of Trainable Parameters: {}'.format(trainable_params))
    print('='*37)
    if optimizer == 'adam':
        optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    elif optimizer == 'sgd':
        optimizer = torch.optim.SGD(model.parameters(), lr=lr, weight_decay=weight_decay)
    elif optimizer == 'momentum':
        optimizer = torch.optim.SGD(model.parameters(), lr=lr, momentum=momentum, weight_decay=weight_decay)
    start_time = time.time()
    
    val_loss, val_acc, val_f1 = get_val_stat(model, val_loader, loss_fn)
    val_f1 = round(val_f1, 4)
    print('>'*5, 'EPOCH 0: validation loss: {}, validation accuracy: {}%, validation f1: {}'.format(0, val_loss, val_acc, val_f1), '<'*5)
        
    epoch_loss_val.append(val_loss)
    epoch_acc_val.append(val_acc)
    epoch_f1_val.append(val_f1)
    epoch_plot.append(num_step)
    for epoch in range(epochs):
        num_step = 0
        print('EPOCH {}/{}'.format(epoch+1, epochs))
        for i, data in enumerate(train_loader):
            optimizer.zero_grad()
            
            batch, y_true = data
            n = batch.shape[0]

            y_pred = model(batch)
            loss = loss_fn(y_pred, y_true)

            loss.backward()
            optimizer.step()
            
            f1 = round(f1_score(y_true, torch.argmax(y_pred, dim=1), average='macro'), 4)
            
            num_correct = (y_true == torch.argmax(y_pred, dim=1)).sum().item()
            acc = round(num_correct*100/n, 4)

            disp_loss = round(loss.item(), 4)
            elapsed_time = round(time.time()-start_time, 2)
            running_samples = n*(i+1)
            
            step_loss_train.append(disp_loss)
            step_acc_train.append(acc)
            step_f1_train.append(f1)
            step_plot.append((epoch+1)*running_samples)
            
            print('({}s) [{}/{}] loss: {}, acc: {}%, f1: {}'.format(elapsed_time, running_samples, N, disp_loss, acc, f1))
            num_step += 1
        val_loss, val_acc, val_f1 = get_val_stat(model, val_loader, loss_fn)
        val_f1 = round(val_f1, 4)
        
        epoch_loss_val.append(val_loss)
        epoch_acc_val.append(val_acc)
        epoch_f1_val.append(val_f1)
        epoch_plot.append(num_step)
        
        print('>'*5, 'EPOCH {}: validation loss: {}, validation accuracy: {}%, validation f1: {}'.format(epoch+1, val_loss, val_acc, val_f1), '<'*5)
    return model, ((step_loss_train, step_acc_train, step_f1_train, step_plot), (epoch_loss_val, epoch_acc_val, epoch_f1_val, epoch_plot))
```

```python
import torch.nn as nn
import torch.nn.functional as F

class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.conv1a = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1)
        self.batchnorm1a = nn.BatchNorm2d(16)
        
        self.conv2a = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1)
        self.batchnorm2a = nn.BatchNorm2d(32)
        
        self.conv3a = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.batchnorm3a = nn.BatchNorm2d(64)
        
        self.conv4a = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.batchnorm4a = nn.BatchNorm2d(128)
        
        self.logits = nn.Linear(128, 10)
        
    def forward(self, inp):
        x = F.relu(self.batchnorm1a(self.conv1a(inp)))
        x = F.relu(self.batchnorm2a(self.conv2a(x)))
        x = F.relu(self.batchnorm3a(self.conv3a(x)))
        x = F.relu(self.batchnorm4a(self.conv4a(x)))
        x = F.adaptive_avg_pool2d(x, (1, 1))
        x = x.view(-1, 128)
        
        out = self.logits(x)
        return out
```

```python
def eval(model, test_loader):
    X_test, y_test = iter(test_loader).next()
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        images, labels = X_test, y_test
        outputs = model(images)
        predicted = torch.argmax(outputs.data, 1)
    total = labels.size(0)
    correct = (predicted == labels).sum().item()
    cm = confusion_matrix(labels.numpy(), predicted)
    cm_norm = normalize_confusion_matrix(cm)
    return correct/total, cm, cm_norm
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd
from seaborn import heatmap
import matplotlib.pyplot as plt

def normalize_confusion_matrix(C):
    return C / C.astype(np.float).sum(axis=1)

def plot_cm(cm, cm_norm):
    df_cm = pd.DataFrame(cm, index = [idx2class[i] for i in range(10)], columns = [idx2class[i] for i in range(10)])
    df_cm_norm = pd.DataFrame(cm_norm, index = [idx2class[i] for i in range(10)], columns = [idx2class[i] for i in range(10)])
    plt.figure(figsize=(25,10))
    
    plt.subplot(121)
    ax1 = heatmap(df_cm, annot=True)
    bottom, top = ax1.get_ylim()
    ax1.set_ylim(bottom + 0.5, top - 0.5)
    plt.title('Confusion Matrix')
    
    plt.subplot(122)
    ax2 = heatmap(df_cm_norm, annot=True)
    bottom, top = ax2.get_ylim()
    ax2.set_ylim(bottom + 0.5, top - 0.5)
    plt.title('Normalized Confusion Matrix')
    plt.show()
```

```python
def plot_history(history):
    S_plot, E_plot = history
    
    plt.figure(figsize=(30,5))
    plt.subplot(131)
    plt.title('Model Loss')
    plt.plot(S_plot[0], label='train_loss')
    plt.plot([157*i for i in range(6)], E_plot[0], linewidth=3, label='val_loss')
    plt.legend()
    plt.subplot(132)
    plt.title('Model Accuracy')
    plt.plot(S_plot[1], label='train_acc')
    plt.plot([157*i for i in range(6)], E_plot[1], linewidth=3, label='val_acc')
    plt.legend()
    plt.subplot(133)
    plt.title('Model F1 Score (macro)')
    plt.plot(S_plot[2], label='train_f1')
    plt.plot([157*i for i in range(6)], E_plot[2], linewidth=3, label='val_f1')
    plt.legend()
    plt.show()
```

```python
cifar_dir='./'
validation_split=0.2
batch_size = 256

train_loader, val_loader, test_loader = get_data(cifar_dir=cifar_dir, batch_size=batch_size, augment=True, validation_split=validation_split)

model = ConvNet()
model.train()
epochs = 5
loss = nn.CrossEntropyLoss()
model, history = train(model, train_loader, val_loader, epochs, loss, batch_size, optimizer='adam', lr=1e-3, weight_decay=5e-4, validation_split=0.2)
```

```python
plot_history(history)
acc, cm, cm_norm = eval(model, test_loader)
plot_cm(cm, cm_norm)
print('Accuracy: {}%'.format(acc*100))
```
